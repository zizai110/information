```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

def CountSecond(time):
    start_time = [3, 19, 0]
    times = time.split(':')
    h = int(times[0]) - start_time[0]
    m = int(times[1]) - start_time[1]
    s = int(times[2]) - start_time[2]
    return h * 3600 + m * 60 + s

def CountSecond2(time):
    times = time.split(':')
    h = int(times[0])
    m = int(times[1])
    s = int(times[2])
    return h * 3600 + m * 60 + s + 74460

def rawData():
    countDict = {item: 0 for item in range(570)}
    fp = open('car_reform.csv', 'r')
    for line in fp:
        rets = line.strip().split(',')
        fmin = int(CountSecond(rets[3]) / 300)

        if fmin in countDict:
            countDict[fmin] += 1
    
    fp = open('car_reform2.csv', 'r')
    for line in fp:
        rets = line.strip().split(',')
        fmin = int(CountSecond2(rets[3]) / 300)

        if fmin in countDict:
            countDict[fmin] += 1
    countList = sorted(countDict.items(), key=lambda x: x[0])
    left, value = [], []
    for item in countList:
        left.append(item[0])
        value.append(item[1])
    return left, value
```

```python
left, value = rawData()
xdata = np.array(left, dtype=np.float32)
ydata = np.array(value, dtype=np.float32)
plt.subplot(2,1,1)
plt.bar(xdata, ydata)
plt.subplot(2,1,2)
sns.kdeplot(ydata, kernel='gau')
```

```python
from pykalman import KalmanFilter
kf = KalmanFilter(initial_state_mean=0, n_dim_obs=1)
measurements = [1,0,1]
y = kf.em(measurements).smooth(ydata)[0]
yplot = [item[0] for item in y]
plt.plot(yplot)
```
